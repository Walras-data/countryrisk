```python
import pandas as pd
import os

world_geo = os.path.join('https://github.com/python-visualization/folium/blob/master/examples/data/world-countries.json')

m = folium.Map(location=[48, -102], zoom_start=3)

folium.Choropleth(
    geo_data=world_geo,
    name='choropleth',
    data=cloropleth,
    columns=['iso_a3', 'pred'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='ExchangeRateRisk'
).add_to(m)

folium.LayerControl().add_to(m)

m
```
