```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import numpy as np
```

```python
base = pd.read_csv('watchlistmay.csv', delimiter = ',')
base
```

```python
base['meanpred'] = base.groupby("year")["pred"].transform("mean")
base
```

```python
import geopandas as gpd 

world = gpd.read_file(
    gpd.datasets.get_path('naturalearth_lowres'))
countrycodes = pd.read_csv('coucodes.csv', delimiter = ';', encoding = "ISO-8859-1")
countrinho = countrycodes[['ISO Code', 'ISO2 Code']]
predictionmap = pd.merge(base, countrinho, left_on = "ISO2 Code", right_on = 'ISO2 Code')
world1 = pd.merge(world, predictionmap, left_on = "iso_a3", right_on = 'ISO Code', how = 'left')
world2 = world1.fillna(0)

ax = world2.plot(column='pred', cmap = 'OrRd', 
                 edgecolor= 'black' ,
                 figsize=(22,15),   
                 scheme='fisher_jenks', 
                 k=10, 
                 legend = True)
#add title to the map
ax.set_title('Exchange Rate Risk', fontdict= 
            {'fontsize':25})
#remove axes
ax.set_axis_off()
#move legend to an empty space
ax.get_legend().set_bbox_to_anchor((.12,.12))


```

```python
#Scaling  and plotting

base["vix"] -= base["vix"].min()
base["vix"] /= base["vix"].max()
base["meanpred"] -= base["meanpred"].min()
base["meanpred"] /= base["meanpred"].max()



```

```python
#Exchange Rate Risk 
### Scaled data shows that we are in the highest risk ever in terms of currency crisis in the following 12 months.

base.plot(x = 'year', y = 'meanpred', figsize = (20,10) )
```

```python
### Index is strongly correlated with VIX 
base.plot(x = 'year', y = ['vix','meanpred'], figsize = (20,10))
```

```python
base['exchangefrente'] = base.groupby('ISO2 Code')['exchange_change_12m'].shift(12)
base
```

```python
basiso = base[base['exchangefrente'] < 1].tail(3000)
basiso
```

```python
plt.figure(figsize=(15,10))
splot = sns.regplot(x="pred", y="exchangefrente",
                    data=basiso,
                    scatter_kws={'alpha':0.15},
                    fit_reg=False)
```

```python
plt.figure(figsize=(16,6))
sns.jointplot("pred", "exchangefrente", 
              scatter_kws={'alpha':0.5},
              data=basiso, 
              kind="reg")
```

```python
plt.figure(figsize=(20,30))
lmplot = sns.lmplot(x='pred', y='exchangefrente', 
                    scatter_kws={'alpha':0.5}, 
                    hue='ISO2 Code', 
                    data=basiso, 
                    fit_reg= False)
```

```python
base['dummycrisis'] = base.groupby("year")["dummy"].transform("sum")
base
```

```python
basinha = base.reset_index()
basinha = basinha.drop_duplicates(['year'])
basinha
```

```python
basinha['aninho'] = basinha['year'].str[:4]
basinha
```

```python
### Number of currency crisis by year 
basisinha = basinha.groupby(['aninho'])["dummycrisis"].sum()
basisinha.plot.bar(x = 'aninho', y = ['dummycrisis'], figsize =(15,10))
```

```python
# Number of currency crisis by month
basinha.plot(x = 'year', y = ['dummycrisis'], figsize =(15,10))
```

```python
### Correlation between model predictions.

matriz = base[["prob1","prob2","prob3","prob4","prob5"]]
corrMatrix = matriz.corr()
mask = np.zeros_like(corrMatrix)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(20, 15))
    ax = sns.heatmap(corrMatrix, mask=mask, vmax=.3, annot = True, cmap="YlGnBu", linewidths = .5, square=True)
```

```python
### Correlation between series
matriz = basiso[["meanpred","vix","oil","NASDAQOMX/NQEM","PERTH/GOLD_USD_D","CHRIS/ICE_DX2","CHRIS/EUREX_FMEM1"]]
corrMatrix = matriz.corr()
mask = np.zeros_like(corrMatrix)
mask[np.triu_indices_from(mask)] = True
with sns.axes_style("white"):
    f, ax = plt.subplots(figsize=(20, 15))
    ax = sns.heatmap(corrMatrix, mask=mask, vmax=.3, annot = True, linewidths = .5, square=True)
```
